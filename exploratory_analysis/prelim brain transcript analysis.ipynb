### Collect data

brain genes is a list of CNS specific genes, from the wilmer database as well as from what I manually curated when we were thinking of doing rt-pcr

```python
wilmer_genes = pd.DataFrame.from_csv('gene_lists/wilmer_brain.txt', sep='\t')['Gene_Symbol']
my_genes = pd.read_excel('gene_lists/brain_transcripts.xlsx', sheetname='Sheet6')['Gene names']

brain_genes=np.unique(np.concatenate([wilmer_genes,my_genes]))
print len(brain_genes)
```

<br />
<br />
This takes the data from cofactor and put into dataframe, and reduce it down to just the reads from CNS specific genes.  Also, the reads were mapped to UCSC transcript isoforms.  For this analysis, I summed all of the reads that came from different transcript isoforms of the same gene.  We can always break out the analysis of different isoforms in the future, but with just three samples, analyzing at the gene level reduces a bit of the noise, but doesn't change any of the overall conclusions.

```python
expression = pd.DataFrame.from_csv('data/NB0001RB_sample_comparative_coverage_master_list_foldval10.txt', sep='\t')
expression=expression[expression['name'].isin(brain_genes)]

#sum reads from same gene
sample_names = ['1585','1585-E','3578','3578-E','9399','9399-E']
samples_exp = pd.DataFrame(expression.groupby('name')[sample_names].sum())

#now just change to more informative column names
sample_list=['SPMS-serum','SPMS-Exo','Active-serum','Active-Exo','Healthy-serum','Healthy-Exo']
samples_exp.columns = sample_list

samples_exp[:5]
```

<br />
<br />
long_exp just puts the data into a format that is a bit easier to plug into some plotting functions in seaborn. Also, since this data is at incredibly different scales, analyzing the log of the data is useful in some situations.  However, we can't take the log of values=0, so I have two different schemes to deal with that.  The first adds a very small amount to all the data, not enough to make a difference, but enough so that we can take a log.  The second just removes all the zero values.  

Adding the small value is probably more accurate, but removing all the zero values creates cleaner data.  So I did both..

```python
long_exp={'expression':[], 'gene':[], 'sample':[], 'prep':[]}
for s in samples_exp.columns.values.tolist():
    for g in samples_exp.index.tolist():
        long_exp['expression'].append(samples_exp[s][g])
        long_exp['gene'].append(g)
        if s[-4:] == '-Exo':
            long_exp['sample'].append(s[:-4])
            long_exp['prep'].append('exosome')
        else:
            long_exp['sample'].append(s[:-6])
            long_exp['prep'].append('circulating RNA')
            
long_exp = pd.DataFrame(long_exp)

#adds infintesimal value to everything
long_exp_inf = long_exp.copy()
long_exp_inf['expression'] = np.log10(long_exp['expression']+.0001)

#gets rid of zero values
long_exp_nozero = long_exp.copy()
long_exp_nozero = long_exp_nozero[long_exp_nozero['expression']>0]
long_exp_nozero['expression'] = np.log10(long_exp_nozero['expression'])



long_exp_inf[:5]
```

### Data Analysis

##### Counting

First I just broadly wanted to look at the total counts for CNS specific genes in each sample, and the total number of CNS specific genes that were detected 

```python
#Add up all the gene counts for CNS specific genes
plt.figure(figsize=(7,5))
sums=[]
for i in sample_list:
    sums.append(samples_exp[i].sum())

plt.figure(num=None, )
width = .25   
fig, ax = plt.subplots()
CircRNA = ax.bar(range(2), sums[:2], width, color=sns.color_palette('deep')[0])
Exo = ax.bar(np.array(range(2))+width, sums[2:4], width, color=sns.color_palette('deep')[1])
Healthy = ax.bar(np.array(range(2))+2*width, sums[4:], width, color=sns.color_palette('deep')[2])

ax.set_xticks(np.array(range(2))+1.75*width)
ax.set_xticklabels( ['circulating RNA', 'Exosome'], size=13 )
ax.set_xlabel('Prep', size=14)
ax.set_ylabel('Sum RPKM', size=14)
ax.legend(['SPMS','Active','Healthy'], fontsize=14,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Total sum of all CNS gene counts', size=16)

plt.show()
print'\n'

#Count how many genes are detected 
counts=[]
for col in sample_list:
    counts.append(len(samples_exp[samples_exp[col]>0]))

plt.figure(num=None,)
width = .25   
fig, ax = plt.subplots()
CircRNA = ax.bar(range(2), counts[:2], width, color=sns.color_palette('deep')[0])
Exo = ax.bar(np.array(range(2))+width, counts[2:4], width, color=sns.color_palette('deep')[1])
Healthy = ax.bar(np.array(range(2))+2*width, counts[4:], width, color=sns.color_palette('deep')[2])

ax.set_xticks(np.array(range(2))+1.75*width)
ax.set_xticklabels( ['circulating RNA', 'Exosome'], size=13 )
ax.set_xlabel('Prep', size=14)
ax.set_ylabel('Number of CNS genes present in sample', size=14)
ax.legend(['SPMS','Active','Healthy'], fontsize=14,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Number of CNS genes with at least one read", size=16)
plt.show()
```

The number of CNS genes with at least one read were relatively consistent across samples.  

These results for total CNS gene counts were pretty interesting.  Just adding up the total amount of CNS genes we see, the exosome preps were relatively similar across samples.  However, when looking at the circulating RNA, there were huge differences - the SPMS was off the charts.  The active patients were (unexpectedly) low.  Not sure if this is reproducible.  In addition, according the cofactor, the samples were normalized to million mapped reads, but there are additional normalization procedures that we could do to account for differences in the amount of reads/RNA in each sample.  Finally, this doesnt even include any age normalization or normalization to brain size that Winston mentioned was super important.  

Overall, it's unclear what the circulating RNA amounts mean, but it was surprising that the exosome amounts were so consistent.

##### Distributions

Next I wanted to look at the distributions of gene counts in each sample.  There were a lot of counts of 0, as well as a huge range in frequency, so I removed the 0's, then take the log for some

```python
def plot_genes(data, title=''):
    plt.figure(figsize=(7,4))
    sns.boxplot(x="prep", y="expression", hue='sample',data=data,whis=np.inf)
    plt.xticks(size=12)
    plt.yticks(size=12)
    plt.ylabel('RPKM',size=14)
    plt.xlabel('Sample',size=14)
    plt.legend(fontsize=14, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.title(title, size=16)
    plt.show()

    
plot_genes(long_exp, title='Expression of individual CNS specific genes')
print'\n'
plot_genes(long_exp_inf, title='Log expression of individual CNS specific genes (small value)')
print'\n'
plot_genes(long_exp_nozero, title='Log expression of individual CNS specific genes (no zeros)')
```

Overall, unless, I remove the 0's, the average is always very very low (0 or .0001).  If I remove the zeros, we start to see some differences between samples, but unclear if it is meaningful.  Particularly in the exosome prep, it seems that while the total counts of CNS genes is about the same across samples (see the plot above), the MS samples have less genes that are all expressed prety highly, whereas the control has more genes that are all expressed at a lower level.  Definitely unclear if this will be reproducible.

##### Composition and correlation

Instead of looking at simple counts of genes, I wanted to look at the composition of CNS genes in each sample, and see if there were similarities or differences.  I started by doing a simple covariance

```python
counts = pd.DataFrame.from_csv('/users/ryan/downloads/grid.csv')
print counts.columns
plt.figure(figsize=(8,6))
plt.title('Gene counts from exosome and serum preps',size=16)
plt.loglog(counts['_1585'],counts['1585_E'], 'o', markersize=3)
plt.xlabel('Gene counts from SPMS serum prep', size=14)
plt.ylabel('Gene counts from SPMS exosome prep', size=14)
plt.axis([.01, 1000000 ,.01 ,1000000])
plt.show()
```

```python
#correlation between circulating RNA prep samples
circ_corr = np.corrcoef(np.array([samples_exp['SPMS-serum'],samples_exp['Active-serum'],samples_exp['Healthy-serum']]))
print 'Circulating RNA correlation coefficients'
print circ_corr,'\n'


plt.figure(figsize=(7,5))
plt.bar(range(3), [circ_corr[1,0],circ_corr[2,0],circ_corr[2,1]], width=.6)
plt.xticks([i+.3 for i in range(3)], ['SPMS-Active','SPMS-Healthy','Active-Healthy'], size=12)
plt.yticks(size=12)
plt.ylabel('Correlation coefficient',size=14)
plt.title('Correlation between circulating RNA samples', size=14)
plt.show()
print'\n\n'


#Correlation between exosome prepped samples
exo_corr = np.corrcoef(np.array([samples_exp['SPMS-Exo'],samples_exp['Active-Exo'],samples_exp['Healthy-Exo']]))
print 'Exosomal RNA correlation coefficients'
print exo_corr,'\n'


plt.figure(figsize=(7,5))
plt.bar(range(3), [exo_corr[1,0],exo_corr[2,0],exo_corr[2,1]],width=.6)
plt.xticks([i+.3 for i in range(3)], ['SPMS-Active','SPMS-Healthy','Active-Healthy'], size=12)
plt.yticks(size=12)
plt.ylabel('Correlation coefficient',size=14)
plt.title('Correlation between exosome RNA samples', size=14)
plt.show()
```

I thought this was pretty striking.  Of course, who knows if it will be reproducible - but the exosome preps between samples were highly correlated - which means that the composition of genes in each was very similar.

However, the composition between the circulating RNA was far less correlated.  The SPMS and Active were the most correlated, and there was far less correlation between either of the MS samples with the healthy controls.
