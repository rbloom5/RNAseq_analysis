```python
def get_sras(in_file, cell_type = None, disease=None):
    sras = {}
    with open(in_file) as f:
        f.readline()
        for line in f:
            if cell_type:
                if line.split('\t')[2] == cell_type:
                    sras[line.split('\t')[0]] = '_'.join(line.split('\t')[:-1])
            elif disease:
                if line.split('\t')[3] == disease:
                    sras[line.split('\t')[0]] = '_'.join(line.split('\t')[:-1])
            else:
                sras[line.split('\t')[0]] = '_'.join(line.split('\t')[:-1])
    return sras

# print get_sras('RNAseq_data_index.txt')            



```

```python
import pandas as pd
import os

def parse_join_htseq(ids):
    all_dfs = []
    
    #loop through ids, download the HTseq file, and put gene counts into dict
    for i in ids:
        count_dict = {}
        htseq_file = i+'_HTseq_counts.txt'
        os.system('aws s3 cp s3://rna-seq/htseq-out/%s .'%htseq_file)
        with open(htseq_file) as f:
            for line in f:
                count_dict[line.split('\t')[0]]=float(line.split('\t')[1].split('\n')[0])
                
        #turn dict into dataframe with SRR id as the index
        count_df = pd.DataFrame.from_dict([count_dict])
        count_df.rename(index={0: ids[i]},inplace=True)
        all_dfs.append(count_df)
        os.system('rm %s'%htseq_file)
    #concatenate all the data frames into one big one
    merged_df = pd.concat(all_dfs)   
    merged_df.fillna(0, inplace=True)
    return merged_df


sra_list = get_sras('RNAseq_data_index.txt')            
all_counts = parse_join_htseq(sra_list)

all_counts.to_csv('all_rnaseq.txt')
        
# parse_join_htseq(['SRR1550997','SRR1550998'])        
        
    
```
